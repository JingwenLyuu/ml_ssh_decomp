```python
import torch
import torch.nn as nn
import numpy as np
import xarray as xr
from torch.utils.data import TensorDataset, DataLoader
from train import apply_inverse_zca_whitening_4d_torch
from unet import UNet
from utils import *
import os

# Setup
setup_random_seeds(42)
device = get_device()

base_path = "gs://leap-persistent/YueWang/SSH/data"
storage_opts = {"token": "cloud", "asynchronous": False}

train = open_zarr(f"{base_path}/train_80_sst.zarr", storage_opts)
test = open_zarr(f"{base_path}/test_80_sst.zarr", storage_opts)
zca = open_zarr(f"{base_path}/zca_80.zarr", storage_opts)

# ZCA transformation parameters
Vt = torch.from_numpy(zca.ubm_Vt.values).float().to(device)
scale = torch.from_numpy(zca.ubm_scale.values).float().to(device)
mean = torch.from_numpy(zca.ubm_mean.values).float().to(device)

# Prepare training data for normalization statistics (SSH only) - same as original code
x_train_ssh_only = torch.from_numpy(train.ssh.values).float().unsqueeze(1).to(device)
x_train_normalized_ssh, min_vals_ssh, max_vals_ssh = min_max_normalize(x_train_ssh_only)

# Prepare SWOT data - following original normalization approach
batch_size = 8

# For strong SSH
x_test_strong_original = torch.from_numpy(strong_ssh.values).float().unsqueeze(0).unsqueeze(0).to(device)
# Apply same normalization as training using SSH normalization parameters
x_test_strong_normalized, _, _ = min_max_normalize(x_test_strong_original, min_vals_ssh, max_vals_ssh)

# Create dummy targets (not used for evaluation but needed for DataLoader)
y_test_dummy = torch.zeros_like(x_test_strong_original).repeat(1, 2, 1, 1)  # 2 channels for compatibility

test_dataset_strong = TensorDataset(x_test_strong_normalized, y_test_dummy)
test_loader_strong = DataLoader(test_dataset_strong, batch_size=batch_size, shuffle=False)

# For weak SSH  
x_test_weak_original = torch.from_numpy(weak_ssh.values).float().unsqueeze(0).unsqueeze(0).to(device)
x_test_weak_normalized, _, _ = min_max_normalize(x_test_weak_original, min_vals_ssh, max_vals_ssh)

test_dataset_weak = TensorDataset(x_test_weak_normalized, y_test_dummy)
test_loader_weak = DataLoader(test_dataset_weak, batch_size=batch_size, shuffle=False)

# Load SSH-only ZCA model
model_ssh_only = UNet(in_channels=1, out_channels=2, initial_features=32, depth=4)
model_ssh_only.to(device)

checkpoint = torch.load('/home/jovyan/GRL_ssh/checkpoints/ssh_input_only.pth', map_location=device)
model_ssh_only.load_state_dict(checkpoint['model_state_dict'])

# Evaluate model on strong SSH
model_ssh_only.eval()
results_swot_strong = {
    'ssh': [], 'ubm_true': [], 'bm_true': [],
    'ubm_pred_mu': [], 'bm_pred_mu': [],
    'ubm_pred_ensembles': [], 'bm_pred_ensembles': []
}

sample_indices_strong = []

with torch.no_grad():
    for i, (batch_x, batch_y) in enumerate(test_loader_strong):
        
        batch_start = i * test_loader_strong.batch_size
        batch_end = min(batch_start + test_loader_strong.batch_size, len(test_dataset_strong))
        current_batch_indices = list(range(batch_start, batch_end))
        sample_indices_strong.extend(current_batch_indices)
            
        batch_x = batch_x.to(device)
        
        outputs = model_ssh_only(batch_x)
        
        ssh_batch_original = x_test_strong_original[current_batch_indices]
        
        # For SWOT data, we don't have true UBM, so we'll set as None or use SSH as placeholder
        ubm_true = ssh_batch_original  # Placeholder - adjust as needed
        bm_true = torch.zeros_like(ssh_batch_original)  # Placeholder - adjust as needed
        
        # Predicted mean in physical space
        mu_zca = outputs[:, 0, ...]
        log_sigma_zca = outputs[:, 1, ...]
        mu_zca_expanded = mu_zca.unsqueeze(1)
        ubm_pred_mu = apply_inverse_zca_whitening_4d_torch(mu_zca_expanded, Vt, scale, mean)
        
        # BM prediction using original SSH scale
        bm_pred_mu = ssh_batch_original - ubm_pred_mu
        
        # Generate ensemble samples
        zca_samples = generate_gaussian_samples(mu_zca, log_sigma_zca, n_samples=30)
        B, n_samples, H, W = zca_samples.shape
        zca_samples_flat = zca_samples.reshape(B * n_samples, 1, H, W)
        ubm_samples_flat = apply_inverse_zca_whitening_4d_torch(zca_samples_flat, Vt, scale, mean)
        ubm_samples = ubm_samples_flat.reshape(B, n_samples, 1, H, W)
        
        # Use original SSH for ensemble BM calculation
        ssh_expanded = ssh_batch_original.unsqueeze(1).expand(-1, n_samples, -1, -1, -1)
        bm_samples = ssh_expanded - ubm_samples
        
        # Store results 
        results_swot_strong['ssh'].append(ssh_batch_original.cpu().numpy())
        results_swot_strong['ubm_true'].append(ubm_true.cpu().numpy())
        results_swot_strong['bm_true'].append(bm_true.cpu().numpy())
        results_swot_strong['ubm_pred_mu'].append(ubm_pred_mu.cpu().numpy())
        results_swot_strong['bm_pred_mu'].append(bm_pred_mu.cpu().numpy())
        results_swot_strong['ubm_pred_ensembles'].append(ubm_samples.cpu().numpy())
        results_swot_strong['bm_pred_ensembles'].append(bm_samples.cpu().numpy())

for key in results_swot_strong:
    results_swot_strong[key] = np.concatenate(results_swot_strong[key], axis=0)

print("SWOT Strong SSH evaluation complete!")

# Evaluate model on weak SSH
results_swot_weak = {
    'ssh': [], 'ubm_true': [], 'bm_true': [],
    'ubm_pred_mu': [], 'bm_pred_mu': [],
    'ubm_pred_ensembles': [], 'bm_pred_ensembles': []
}

sample_indices_weak = []

with torch.no_grad():
    for i, (batch_x, batch_y) in enumerate(test_loader_weak):
        
        batch_start = i * test_loader_weak.batch_size
        batch_end = min(batch_start + test_loader_weak.batch_size, len(test_dataset_weak))
        current_batch_indices = list(range(batch_start, batch_end))
        sample_indices_weak.extend(current_batch_indices)
            
        batch_x = batch_x.to(device)
        
        outputs = model_ssh_only(batch_x)
        
        ssh_batch_original = x_test_weak_original[current_batch_indices]
        
        # For SWOT data, we don't have true UBM, so we'll set as None or use SSH as placeholder
        ubm_true = ssh_batch_original  # Placeholder - adjust as needed
        bm_true = torch.zeros_like(ssh_batch_original)  # Placeholder - adjust as needed
        
        # Predicted mean in physical space
        mu_zca = outputs[:, 0, ...]
        log_sigma_zca = outputs[:, 1, ...]
        mu_zca_expanded = mu_zca.unsqueeze(1)
        ubm_pred_mu = apply_inverse_zca_whitening_4d_torch(mu_zca_expanded, Vt, scale, mean)
        
        # BM prediction using original SSH scale
        bm_pred_mu = ssh_batch_original - ubm_pred_mu
        
        # Generate ensemble samples
        zca_samples = generate_gaussian_samples(mu_zca, log_sigma_zca, n_samples=30)
        B, n_samples, H, W = zca_samples.shape
        zca_samples_flat = zca_samples.reshape(B * n_samples, 1, H, W)
        ubm_samples_flat = apply_inverse_zca_whitening_4d_torch(zca_samples_flat, Vt, scale, mean)
        ubm_samples = ubm_samples_flat.reshape(B, n_samples, 1, H, W)
        
        # Use original SSH for ensemble BM calculation
        ssh_expanded = ssh_batch_original.unsqueeze(1).expand(-1, n_samples, -1, -1, -1)
        bm_samples = ssh_expanded - ubm_samples
        
        # Store results 
        results_swot_weak['ssh'].append(ssh_batch_original.cpu().numpy())
        results_swot_weak['ubm_true'].append(ubm_true.cpu().numpy())
        results_swot_weak['bm_true'].append(bm_true.cpu().numpy())
        results_swot_weak['ubm_pred_mu'].append(ubm_pred_mu.cpu().numpy())
        results_swot_weak['bm_pred_mu'].append(bm_pred_mu.cpu().numpy())
        results_swot_weak['ubm_pred_ensembles'].append(ubm_samples.cpu().numpy())
        results_swot_weak['bm_pred_ensembles'].append(bm_samples.cpu().numpy())

for key in results_swot_weak:
    results_swot_weak[key] = np.concatenate(results_swot_weak[key], axis=0)

print("SWOT Weak SSH evaluation complete!")

# Create xarray datasets following original pattern
swot_results = [
    ('swot_strong', results_swot_strong, True, False),
    ('swot_weak', results_swot_weak, True, False)
]

# Store all datasets
eval_datasets_swot = {}

for model_name, results, has_ensembles, has_sst in swot_results:
    print(f"Creating dataset for {model_name}...")
    
    eval_dataset = create_evaluation_dataset(results, model_name, has_ensembles, has_sst)
    
    # Store the dataset
    eval_datasets_swot[model_name] = eval_dataset

print("All SWOT evaluations complete!")
print("Available datasets:", list(eval_datasets_swot.keys()))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
import matplotlib.gridspec as gridspec
import matplotlib.ticker as ticker

import cmocean.cm as cmo
speed_cmap = cmo.thermal  # For speed fields (as specified in your code)
scalar_cmap = cmo.ice  # For SSH fields (as specified in your code)

# Set professional matplotlib parameters for publication
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'DejaVu Sans', 'Helvetica', 'sans-serif'],
    'mathtext.fontset': 'stix',
    'axes.grid': False,
    'figure.facecolor': 'white',
    'axes.facecolor': 'white'
})

# Define Coriolis parameter for Agulhas region (approximately 40°S)
f = -9.37e-5  # s^-1 (negative in Southern Hemisphere)
g = 9.81  # gravitational acceleration (m/s^2)

def calculate_geostrophic_velocities(eta, f, g, dx, dy):
    """Calculate geostrophic velocities from SSH."""
    # Calculate gradients
    deta_dx = np.zeros_like(eta)
    deta_dy = np.zeros_like(eta)
    
    # x-gradient (central differences)
    deta_dx[:, 1:-1] = (eta[:, 2:] - eta[:, :-2]) / (2 * dx)
    # y-gradient (central differences)
    deta_dy[1:-1, :] = (eta[2:, :] - eta[:-2, :]) / (2 * dy)
    
    # Handle boundaries
    deta_dx[:, 0] = (eta[:, 1] - eta[:, 0]) / dx
    deta_dx[:, -1] = (eta[:, -1] - eta[:, -2]) / dx
    deta_dy[0, :] = (eta[1, :] - eta[0, :]) / dy
    deta_dy[-1, :] = (eta[-1, :] - eta[-2, :]) / dy
    
    # Calculate velocities
    u = -(g/f) * deta_dy
    v = (g/f) * deta_dx
    
    return u, v

def calculate_flow_properties(u, v, dx, dy):
    """Calculate vorticity and strain from velocity components."""
    # Calculate derivatives
    du_dx = np.zeros_like(u)
    du_dy = np.zeros_like(u)
    dv_dx = np.zeros_like(v)
    dv_dy = np.zeros_like(v)
    
    # Interior points using central differences
    du_dx[:, 1:-1] = (u[:, 2:] - u[:, :-2]) / (2 * dx)
    du_dy[1:-1, :] = (u[2:, :] - u[:-2, :]) / (2 * dy)
    dv_dx[:, 1:-1] = (v[:, 2:] - v[:, :-2]) / (2 * dx)
    dv_dy[1:-1, :] = (v[2:, :] - v[:-2, :]) / (2 * dy)
    
    # Handle boundaries
    du_dx[:, 0] = (u[:, 1] - u[:, 0]) / dx
    du_dx[:, -1] = (u[:, -1] - u[:, -2]) / dx
    du_dy[0, :] = (u[1, :] - u[0, :]) / dy
    du_dy[-1, :] = (u[-1, :] - u[-2, :]) / dy
    dv_dx[:, 0] = (v[:, 1] - v[:, 0]) / dx
    dv_dx[:, -1] = (v[:, -1] - v[:, -2]) / dx
    dv_dy[0, :] = (v[1, :] - v[0, :]) / dy
    dv_dy[-1, :] = (v[-1, :] - v[-2, :]) / dy
    
    # Vorticity
    vorticity = dv_dx - du_dy
    
    # Strain components
    strain_normal = du_dx - dv_dy
    strain_shear = dv_dx + du_dy
    
    # Total strain magnitude
    strain = np.sqrt(strain_normal**2 + strain_shear**2)
    
    return vorticity, strain

def format_axes_professional(ax, title, dx_km=1.5):
    """Format axes."""
    # Assuming 80x80 grid where each pixel is 1.5km
    size = 80
    
    # Calculate tick positions - show every 20 pixels (30 km)
    tick_positions = np.arange(0, size+1, 20)  # 0, 20, 40, 60, 80
    km_ticks = tick_positions * dx_km  # Convert to km
    
    # Set ticks and labels
    ax.set_xticks(tick_positions)
    ax.set_yticks(tick_positions)
    ax.set_xticklabels([f'{int(km)}' for km in km_ticks], fontsize=20)
    ax.set_yticklabels([f'{int(km)}' for km in km_ticks], fontsize=20)
    
    # Style ticks
    ax.tick_params(axis='both', which='major', length=6, width=1.5, labelsize=20)
    
    # Set title with proper spacing
    ax.set_title(title, fontsize=28, pad=15, fontweight='bold')
    
    # Add axis labels only for bottom row and left column
    return ax

def add_professional_colorbar(fig, ax, im, label, extend='neither', labelpad=1, data_type='ssh_strong'):
    """Add a professional colorbar with manual scientific notation formatting."""
    cbar = fig.colorbar(im, ax=ax, pad=0.02, fraction=0.06, extend=extend, aspect=17, shrink=0.93)
    cbar.set_label(label, fontsize=18, labelpad=labelpad)
    cbar.ax.tick_params(labelsize=18, length=5, width=1.5)
    
    # Manually set scientific notation labels based on data type
    if data_type == 'ssh_strong':  # SSH and BM strong: range -1 to 1
        ticks = [-1.0, -0.5, 0.0, 0.5, 1.0]
        labels = ['-1.0', '-0.5', '0.0', '0.5', '1.0']
        cbar.set_ticks(ticks)
        cbar.set_ticklabels(labels)
        
    elif data_type == 'ubm_strong':  # UBM: range -0.01 to 0.01
        ticks = [-0.01, -0.005, 0.0, 0.005, 0.01]
        labels = ['-1.0', '-0.5', '0.0', '0.5', '1.0']  # Just coefficients
        cbar.set_ticks(ticks)
        cbar.set_ticklabels(labels)
        # Manually add ×10⁻² label above colorbar
        cbar.ax.text(-0.2, 1.045, '×10⁻²', transform=cbar.ax.transAxes,
                   fontsize=14, ha='left', va='bottom')
        
    elif data_type == 'speed_strong':  # Speed strong: range 0 to 1
        ticks = [0.0, 0.25, 0.5, 0.75, 1.0]
        labels = ['0.0', '0.25', '0.5', '0.75', '1.0']
        cbar.set_ticks(ticks)
        cbar.set_ticklabels(labels)
        
    elif data_type == 'speed_ubm_strong':  # UBM Speed: range 0 to 0.2
        ticks = [0.0, 0.05, 0.1, 0.15, 0.2]
        labels = ['0.0', '0.05', '0.1', '0.15', '0.2']
        cbar.set_ticks(ticks)
        cbar.set_ticklabels(labels)
    
    return cbar

# Define grid spacing
dx = 1500  # meters
dy = 1500  # meters
dx_km = dx / 1000  # 1.5 km

# Calculate for SSH
u_ssh_strong, v_ssh_strong = calculate_geostrophic_velocities(strong_ssh_sample, f, g, dx, dy)
speed_ssh_strong = np.sqrt(u_ssh_strong**2 + v_ssh_strong**2)
vort_ssh_strong, strain_ssh_strong = calculate_flow_properties(u_ssh_strong, v_ssh_strong, dx, dy)

# Calculate for BM
u_bm_strong, v_bm_strong = calculate_geostrophic_velocities(strong_bm_sample, f, g, dx, dy)
speed_bm_strong = np.sqrt(u_bm_strong**2 + v_bm_strong**2)
vort_bm_strong, strain_bm_strong = calculate_flow_properties(u_bm_strong, v_bm_strong, dx, dy)

# Calculate for UBM
u_ubm_strong, v_ubm_strong = calculate_geostrophic_velocities(strong_ubm_sample, f, g, dx, dy)
speed_ubm_strong = np.sqrt(u_ubm_strong**2 + v_ubm_strong**2)
vort_ubm_strong, strain_ubm_strong = calculate_flow_properties(u_ubm_strong, v_ubm_strong, dx, dy)

# Create figure
fig = plt.figure(figsize=(15, 9))
gs = gridspec.GridSpec(2, 3, hspace=0.35, wspace=0.5, 
                       left=0.08, right=0.95, top=0.92, bottom=0.08)

# Row 0: SSH/Eta fields
ax00 = fig.add_subplot(gs[0, 0])
ax01 = fig.add_subplot(gs[0, 1]) 
ax02 = fig.add_subplot(gs[0, 2])

# Row 1: Speed fields
ax10 = fig.add_subplot(gs[1, 0])
ax11 = fig.add_subplot(gs[1, 1])
ax12 = fig.add_subplot(gs[1, 2])

# Panel (a): SSH
im00 = ax00.imshow(strong_ssh_sample, cmap=scalar_cmap, vmin=-1, vmax=1, 
                   extent=[0, 80, 0, 80])
format_axes_professional(ax00, 'Raw SSH (SWOT)', dx_km)

# Panel (b): BM
im01 = ax01.imshow(strong_bm_sample, cmap=scalar_cmap, vmin=-1, vmax=1,
                   extent=[0, 80, 0, 80])
format_axes_professional(ax01, 'BM Pred Mean', dx_km)

# Panel (c): UBM
im02 = ax02.imshow(strong_ubm_sample, cmap=scalar_cmap, vmin=-0.01, vmax=0.01,
                   extent=[0, 80, 0, 80])
format_axes_professional(ax02, 'UBM Pred Mean', dx_km)

# Panel (d): Speed SSH
im10 = ax10.imshow(speed_ssh_strong, cmap=speed_cmap, vmin=0, vmax=1,
                   extent=[0, 80, 0, 80])
format_axes_professional(ax10, 'SSH GSpeed', dx_km)

# Panel (e): Speed BM
im11 = ax11.imshow(speed_bm_strong, cmap=speed_cmap, vmin=0, vmax=1,
                   extent=[0, 80, 0, 80])
format_axes_professional(ax11, 'BM GSpeed', dx_km)

# Panel (f): Speed UBM
im12 = ax12.imshow(speed_ubm_strong, cmap=speed_cmap, vmin=0, vmax=0.2,
                   extent=[0, 80, 0, 80])
format_axes_professional(ax12, 'UBM GSpeed', dx_km)

# Add axis labels for bottom row
ax10.set_xlabel('Distance (km)', fontsize=20)
ax11.set_xlabel('Distance (km)', fontsize=20)
ax12.set_xlabel('Distance (km)', fontsize=20)

# Add axis labels for left column
ax00.set_ylabel('Distance (km)', fontsize=20)
ax10.set_ylabel('Distance (km)', fontsize=20)

# Add colorbars with appropriate ranges for strong data
# First row (SSH fields)
cbar00 = add_professional_colorbar(fig, ax00, im00, '(m)', extend='both', labelpad=3, data_type='ssh_strong')
cbar01 = add_professional_colorbar(fig, ax01, im01, '(m)', extend='both', labelpad=3, data_type='ssh_strong') 
cbar02 = add_professional_colorbar(fig, ax02, im02, '(m)', extend='both', labelpad=3, data_type='ubm_strong')

# Second row (Speed fields)
cbar10 = add_professional_colorbar(fig, ax10, im10, '(m/s)', extend='max', labelpad=3, data_type='speed_strong')
cbar11 = add_professional_colorbar(fig, ax11, im11, '(m/s)', extend='max', labelpad=3, data_type='speed_strong')
cbar12 = add_professional_colorbar(fig, ax12, im12, '(m/s)', extend='max', labelpad=3, data_type='speed_ubm_strong')

# Save
plt.savefig('figures/Figure_3/strong_ssh_professional.png', 
           dpi=300, bbox_inches='tight', facecolor='white', 
           edgecolor='none', format='png')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# Professional publication settings
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'DejaVu Sans', 'Helvetica', 'sans-serif'],
    'mathtext.fontset': 'stix',  
    'axes.grid': False,
    'figure.facecolor': 'white',
    'axes.facecolor': 'white', 
    'text.usetex': False,  # Ensure consistent text rendering
})

# Professional color palette
colors = {
    'ssh': '#1f77b4',      # Blue
    'bm': '#d62728',       # Red  
    'ubm': '#2ca02c',      # Green
    'reference': '#000000' # Black
}

# Define a function to calculate the PSD if it's not already defined
def calculate_psd_km(field, dx_km=1.5):
    """
    Calculate the 2D Power Spectral Density and convert to 1D radial average.
    
    Parameters:
    -----------
    field : 2D numpy array
        The input field (e.g., SSH, BM, UBM)
    dx_km : float
        Grid spacing in kilometers
        
    Returns:
    --------
    psd_1d : numpy array
        1D radially averaged power spectrum
    freq_r : numpy array
        Corresponding radial frequencies in cycles/km
    """
    # Get dimensions
    ny, nx = field.shape
    
    # Apply Hann window to reduce edge effects
    window = np.outer(np.hanning(ny), np.hanning(nx))
    field_windowed = field * window
    
    # Compute 2D FFT and shift zero frequency to center
    fft2 = np.fft.fft2(field_windowed)
    fft2_shifted = np.fft.fftshift(fft2)
    
    # Calculate the Power Spectral Density
    psd2d = np.abs(fft2_shifted)**2
    
    # Normalize by number of points and grid spacing
    psd2d = psd2d / (nx * ny) * (dx_km**2)
    
    # Create frequency grids
    freq_x = np.fft.fftshift(np.fft.fftfreq(nx, d=dx_km))
    freq_y = np.fft.fftshift(np.fft.fftfreq(ny, d=dx_km))
    
    # Create meshgrid for radial binning
    kx, ky = np.meshgrid(freq_x, freq_y)
    k_rad = np.sqrt(kx**2 + ky**2)
    
    # Create radial bins
    max_freq = np.max(k_rad)
    n_bins = min(nx, ny) // 2
    freq_bins = np.linspace(0, max_freq, n_bins+1)
    
    # Calculate bin centers
    freq_r = (freq_bins[1:] + freq_bins[:-1]) / 2
    
    # Initialize output array
    psd_1d = np.zeros(n_bins)
    
    # Radial binning
    for i in range(n_bins):
        bin_mask = (k_rad >= freq_bins[i]) & (k_rad < freq_bins[i+1])
        if np.any(bin_mask):
            psd_1d[i] = np.mean(psd2d[bin_mask])
    
    return psd_1d, freq_r

def calculate_ensemble_psd_envelope(ensemble_samples, dx_km=1.5):
    """
    Calculate PSD envelope (min/max/percentiles) from ensemble samples
    
    Parameters:
    -----------
    ensemble_samples : numpy array
        Shape: (n_samples, height, width) - ensemble predictions
    dx_km : float
        Grid spacing in kilometers
    
    Returns:
    --------
    psd_envelope : dict
        Dictionary containing 'min', 'max', 'p25', 'p75', 'median' PSDs
    freq_r : numpy array
        Corresponding radial frequencies
    """
    n_samples = ensemble_samples.shape[0]
    
    # Calculate PSD for each ensemble member
    psd_list = []
    for i in range(n_samples):
        psd_1d, freq_r = calculate_psd_km(ensemble_samples[i], dx_km)
        psd_list.append(psd_1d)
    
    # Stack all PSDs
    psd_array = np.stack(psd_list, axis=0)  # Shape: (n_samples, n_freq)
    
    # Calculate envelope statistics
    psd_envelope = {
        'min': np.min(psd_array, axis=0),
        'max': np.max(psd_array, axis=0),
        'p25': np.percentile(psd_array, 25, axis=0),
        'p75': np.percentile(psd_array, 75, axis=0),
        'median': np.median(psd_array, axis=0)
    }
    
    return psd_envelope, freq_r

# Extract ensemble samples from evaluation results
ubm_samples_strong = eval_datasets_swot['swot_strong']['ubm_pred_samples'].values[0]  # Shape: (30, H, W)
bm_samples_strong = eval_datasets_swot['swot_strong']['bm_pred_samples'].values[0]   # Shape: (30, H, W)

ubm_samples_weak = eval_datasets_swot['swot_weak']['ubm_pred_samples'].values[0]     # Shape: (30, H, W)
bm_samples_weak = eval_datasets_swot['swot_weak']['bm_pred_samples'].values[0]       # Shape: (30, H, W)

# Create the figure with 1x2 subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# ===== LEFT SUBPLOT: STRONG SSH CASE =====
ax1 = axes[0]

# Extract SSH field for strong case (original input)
ssh_strong = strong_ssh_sample

# Calculate PSD for original SSH
psd_ssh_strong, freq_r_ssh = calculate_psd_km(ssh_strong)

# Calculate ensemble PSD envelopes
ubm_envelope_strong, freq_r_ubm = calculate_ensemble_psd_envelope(ubm_samples_strong)
bm_envelope_strong, freq_r_bm = calculate_ensemble_psd_envelope(bm_samples_strong)

# Plot SSH (reference)
ax1.plot(freq_r_ssh, psd_ssh_strong, 
         color=colors['ssh'], linewidth=3.5, label='SSH',
         solid_capstyle='round')

# Plot UBM uncertainty envelope
ax1.fill_between(freq_r_ubm, ubm_envelope_strong['min'], ubm_envelope_strong['max'],
                 color=colors['ubm'], alpha=0.4, label='UBM envelope')
# ax1.fill_between(freq_r_ubm, ubm_envelope_strong['p25'], ubm_envelope_strong['p75'],
#                  color=colors['ubm'], alpha=0.4)

# Plot BM uncertainty envelope  
ax1.fill_between(freq_r_bm, bm_envelope_strong['min'], bm_envelope_strong['max'],
                 color=colors['bm'], alpha=0.4, label='BM envelope')
# ax1.fill_between(freq_r_bm, bm_envelope_strong['p25'], bm_envelope_strong['p75'],
#                  color=colors['bm'], alpha=0.4)

# Professional plot customization
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlim(8e-3, 5e-1)
ax1.set_ylim(1e-14, 1e1)

# Professional tick styling
ax1.tick_params(axis='both', which='major', labelsize=22, length=8, 
               width=1.5, direction='in') 
ax1.tick_params(axis='both', which='minor', length=4, 
               width=0.8, direction='in')   

# Panel label
ax1.text(0.02, 0.98, 'a)', transform=ax1.transAxes, 
        fontsize=24, fontweight='bold', va='top', ha='left')

# Labels
ax1.set_xlabel(r'Wavenumber (cpkm)', fontsize=24, fontweight='normal')
ax1.set_ylabel(r'PSD (m$^2$ cpkm$^{-1}$)', fontsize=24, fontweight='normal')

# ===== RIGHT SUBPLOT: WEAK SSH CASE =====
ax2 = axes[1]

# Extract SSH field for weak case
ssh_weak = weak_ssh_sample

# Calculate PSD for original SSH
psd_ssh_weak, freq_r_ssh_weak = calculate_psd_km(ssh_weak)

# Calculate ensemble PSD envelopes
ubm_envelope_weak, freq_r_ubm_weak = calculate_ensemble_psd_envelope(ubm_samples_weak)
bm_envelope_weak, freq_r_bm_weak = calculate_ensemble_psd_envelope(bm_samples_weak)

# Plot SSH (reference)
ax2.plot(freq_r_ssh_weak, psd_ssh_weak, 
         color=colors['ssh'], linewidth=3.5, label='SSH',
         solid_capstyle='round')

# Plot UBM uncertainty envelope
ax2.fill_between(freq_r_ubm_weak, ubm_envelope_weak['min'], ubm_envelope_weak['max'],
                 color=colors['ubm'], alpha=0.4, label='UBM envelope')
# ax2.fill_between(freq_r_ubm_weak, ubm_envelope_weak['p25'], ubm_envelope_weak['p75'],
#                  color=colors['ubm'], alpha=0.4)

# Plot BM uncertainty envelope
ax2.fill_between(freq_r_bm_weak, bm_envelope_weak['min'], bm_envelope_weak['max'],
                 color=colors['bm'], alpha=0.4, label='BM envelope')
# ax2.fill_between(freq_r_bm_weak, bm_envelope_weak['p25'], bm_envelope_weak['p75'],
#                  color=colors['bm'], alpha=0.4)

# Professional plot customization
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlim(8e-3, 5e-1)
ax2.set_ylim(1e-14, 1e1)

# Professional tick styling
ax2.tick_params(axis='both', which='major', labelsize=22, length=8, 
               width=1.5, direction='in') 
ax2.tick_params(axis='both', which='minor', length=4, 
               width=0.8, direction='in')   

# Hide y-axis labels for right subplot
ax2.tick_params(axis='y', labelleft=False)

# Panel label
ax2.text(0.02, 0.98, 'b)', transform=ax2.transAxes, 
        fontsize=24, fontweight='bold', va='top', ha='left')

# Labels
ax2.set_xlabel(r'Wavenumber (cpkm)', fontsize=24, fontweight='normal')

# Create professional legend
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, -0.1), 
           ncol=3, frameon=True, fontsize=22, columnspacing=3.5,
           fancybox=True, shadow=False, framealpha=0.9,
           edgecolor='black', facecolor='white')

plt.tight_layout()
plt.subplots_adjust(bottom=0.2)  # Space for legend

# Professional save options
plt.savefig('figures/Figure_3/psd_uncertainty_envelope.png', 
            bbox_inches='tight', dpi=300, facecolor='white', 
            edgecolor='none', format='png')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

# Professional publication settings
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'DejaVu Sans', 'Helvetica', 'sans-serif'],
    'mathtext.fontset': 'stix',  
    'axes.grid': False,
    'figure.facecolor': 'white',
    'axes.facecolor': 'white', 
    'text.usetex': False,  # Ensure consistent text rendering
})

# Professional color palette
colors = {
    'ssh': '#1f77b4',      # Blue
    'bm': '#d62728',       # Red  
    'ubm': '#2ca02c',      # Green
    'reference': '#000000' # Black
}

# Define a function to calculate the PSD if it's not already defined
def calculate_psd_km(field, dx_km=1.5):
    """
    Calculate the 2D Power Spectral Density and convert to 1D radial average.
    
    Parameters:
    -----------
    field : 2D numpy array
        The input field (e.g., SSH, BM, UBM)
    dx_km : float
        Grid spacing in kilometers
        
    Returns:
    --------
    psd_1d : numpy array
        1D radially averaged power spectrum
    freq_r : numpy array
        Corresponding radial frequencies in cycles/km
    """
    # Get dimensions
    ny, nx = field.shape
    
    # Apply Hann window to reduce edge effects
    window = np.outer(np.hanning(ny), np.hanning(nx))
    field_windowed = field * window
    
    # Compute 2D FFT and shift zero frequency to center
    fft2 = np.fft.fft2(field_windowed)
    fft2_shifted = np.fft.fftshift(fft2)
    
    # Calculate the Power Spectral Density
    psd2d = np.abs(fft2_shifted)**2
    
    # Normalize by number of points and grid spacing
    psd2d = psd2d / (nx * ny) * (dx_km**2)
    
    # Create frequency grids
    freq_x = np.fft.fftshift(np.fft.fftfreq(nx, d=dx_km))
    freq_y = np.fft.fftshift(np.fft.fftfreq(ny, d=dx_km))
    
    # Create meshgrid for radial binning
    kx, ky = np.meshgrid(freq_x, freq_y)
    k_rad = np.sqrt(kx**2 + ky**2)
    
    # Create radial bins
    max_freq = np.max(k_rad)
    n_bins = min(nx, ny) // 2
    freq_bins = np.linspace(0, max_freq, n_bins+1)
    
    # Calculate bin centers
    freq_r = (freq_bins[1:] + freq_bins[:-1]) / 2
    
    # Initialize output array
    psd_1d = np.zeros(n_bins)
    
    # Radial binning
    for i in range(n_bins):
        bin_mask = (k_rad >= freq_bins[i]) & (k_rad < freq_bins[i+1])
        if np.any(bin_mask):
            psd_1d[i] = np.mean(psd2d[bin_mask])
    
    return psd_1d, freq_r

# Create the figure with 1x2 subplots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# ===== LEFT SUBPLOT: STRONG SSH CASE =====
ax1 = axes[0]

# Extract fields for strong case
ssh_strong = strong_ssh_sample
bm_strong = strong_bm_sample
ubm_strong = strong_ubm_sample

# Calculate PSDs for strong case
psd_ssh_strong, freq_r_ssh_strong = calculate_psd_km(ssh_strong)
psd_bm_strong, freq_r_bm_strong = calculate_psd_km(bm_strong)
psd_ubm_strong, freq_r_ubm_strong = calculate_psd_km(ubm_strong)

# Plot with professional styling
ax1.plot(freq_r_ssh_strong, psd_ssh_strong, 
         color=colors['ssh'], linewidth=3.5, label='SSH',
         solid_capstyle='round')
ax1.plot(freq_r_bm_strong, psd_bm_strong, 
         color=colors['bm'], linewidth=3.5, label='BM',
         solid_capstyle='round')
ax1.plot(freq_r_ubm_strong, psd_ubm_strong, 
         color=colors['ubm'], linewidth=3.5, label='UBM',
         solid_capstyle='round')

# Professional plot customization
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlim(8e-3, 5e-1)
ax1.set_ylim(1e-14, 1e1)  # Same y-axis for both subplots

# Professional tick styling
ax1.tick_params(axis='both', which='major', labelsize=22, length=8, 
               width=1.5, direction='in') 
ax1.tick_params(axis='both', which='minor', length=4, 
               width=0.8, direction='in')   

# Panel label
ax1.text(0.02, 0.98, 'a)', transform=ax1.transAxes, 
        fontsize=24, fontweight='bold', va='top', ha='left')

# Labels with proper formatting
ax1.set_xlabel(r'Wavenumber (cpkm)', fontsize=24, fontweight='normal')
ax1.set_ylabel(r'PSD (m$^2$ cpkm$^{-1}$)', fontsize=24, fontweight='normal')

# ===== RIGHT SUBPLOT: WEAK SSH CASE =====
ax2 = axes[1]

# Extract fields for weak case
ssh_weak = weak_ssh_sample
bm_weak = weak_bm_sample
ubm_weak = weak_ubm_sample

# Calculate PSDs for weak case
psd_ssh_weak, freq_r_ssh_weak = calculate_psd_km(ssh_weak)
psd_bm_weak, freq_r_bm_weak = calculate_psd_km(bm_weak)
psd_ubm_weak, freq_r_ubm_weak = calculate_psd_km(ubm_weak)

# Plot with professional styling
ax2.plot(freq_r_ssh_weak, psd_ssh_weak, 
         color=colors['ssh'], linewidth=3.5, label='SSH',
         solid_capstyle='round')
ax2.plot(freq_r_bm_weak, psd_bm_weak, 
         color=colors['bm'], linewidth=3.5, label='BM',
         solid_capstyle='round')
ax2.plot(freq_r_ubm_weak, psd_ubm_weak, 
         color=colors['ubm'], linewidth=3.5, label='UBM',
         solid_capstyle='round')

# Professional plot customization
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlim(8e-3, 5e-1)
ax2.set_ylim(1e-14, 1e1)  # Same y-axis for both subplots

# Professional tick styling
ax2.tick_params(axis='both', which='major', labelsize=22, length=8, 
               width=1.5, direction='in') 
ax2.tick_params(axis='both', which='minor', length=4, 
               width=0.8, direction='in')   

# Hide y-axis labels for right subplot
ax2.tick_params(axis='y', labelleft=False)

# Panel label
ax2.text(0.02, 0.98, 'b)', transform=ax2.transAxes, 
        fontsize=24, fontweight='bold', va='top', ha='left')

# Labels with proper formatting
ax2.set_xlabel(r'Wavenumber (cpkm)', fontsize=24, fontweight='normal')

# Create professional legend
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, -0.1), 
           ncol=3, frameon=True, fontsize=22, columnspacing=3.5,
           fancybox=True, shadow=False, framealpha=0.9,
           edgecolor='black', facecolor='white')

plt.tight_layout()
plt.subplots_adjust(bottom=0.2)  # Space for legend

# Professional save options
plt.savefig('figures/Figure_3/psd_comparison_strong_weak.png', 
            bbox_inches='tight', dpi=300, facecolor='white', 
            edgecolor='none', format='png')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
import cmocean.cm as cmo  

# Define Coriolis parameter for Algulhas region (approximately 40°S)
f = -9.37e-5  # s^-1 (negative in Southern Hemisphere)
g = 9.81  # gravitational acceleration (m/s^2)

# Calculate SSH velocities (if not already calculated)
def calculate_geostrophic_velocities(eta, f, g, dx, dy):
    """Calculate geostrophic velocities from SSH."""
    # Calculate gradients
    deta_dx = np.zeros_like(eta)
    deta_dy = np.zeros_like(eta)
    
    # x-gradient (central differences)
    deta_dx[:, 1:-1] = (eta[:, 2:] - eta[:, :-2]) / (2 * dx)
    # y-gradient (central differences)
    deta_dy[1:-1, :] = (eta[2:, :] - eta[:-2, :]) / (2 * dy)
    
    # Handle boundaries
    deta_dx[:, 0] = (eta[:, 1] - eta[:, 0]) / dx
    deta_dx[:, -1] = (eta[:, -1] - eta[:, -2]) / dx
    deta_dy[0, :] = (eta[1, :] - eta[0, :]) / dy
    deta_dy[-1, :] = (eta[-1, :] - eta[-2, :]) / dy
    
    # Calculate velocities
    u = -(g/f) * deta_dy
    v = (g/f) * deta_dx
    
    return u, v

def calculate_flow_properties(u, v, dx, dy):
    """Calculate vorticity and strain from velocity components."""
    # Calculate derivatives
    du_dx = np.zeros_like(u)
    du_dy = np.zeros_like(u)
    dv_dx = np.zeros_like(v)
    dv_dy = np.zeros_like(v)
    
    # Interior points using central differences
    du_dx[:, 1:-1] = (u[:, 2:] - u[:, :-2]) / (2 * dx)
    du_dy[1:-1, :] = (u[2:, :] - u[:-2, :]) / (2 * dy)
    dv_dx[:, 1:-1] = (v[:, 2:] - v[:, :-2]) / (2 * dx)
    dv_dy[1:-1, :] = (v[2:, :] - v[:-2, :]) / (2 * dy)
    
    # Handle boundaries
    du_dx[:, 0] = (u[:, 1] - u[:, 0]) / dx
    du_dx[:, -1] = (u[:, -1] - u[:, -2]) / dx
    du_dy[0, :] = (u[1, :] - u[0, :]) / dy
    du_dy[-1, :] = (u[-1, :] - u[-2, :]) / dy
    dv_dx[:, 0] = (v[:, 1] - v[:, 0]) / dx
    dv_dx[:, -1] = (v[:, -1] - v[:, -2]) / dx
    dv_dy[0, :] = (v[1, :] - v[0, :]) / dy
    dv_dy[-1, :] = (v[-1, :] - v[-2, :]) / dy
    
    # Vorticity
    vorticity = dv_dx - du_dy
    
    # Strain components
    strain_normal = du_dx - dv_dy
    strain_shear = dv_dx + du_dy
    
    # Total strain magnitude
    strain = np.sqrt(strain_normal**2 + strain_shear**2)
    
    return vorticity, strain

# Define grid spacing
dx = 1500  # meters
dy = 1500  # meters

# Calculate for SSH
u_ssh, v_ssh = calculate_geostrophic_velocities(strong_ssh_sample, f, g, dx, dy)
speed_ssh = np.sqrt(u_ssh**2 + v_ssh**2)
vort_ssh, strain_ssh = calculate_flow_properties(u_ssh, v_ssh, dx, dy)

# Calculate for BM
u_bm, v_bm = calculate_geostrophic_velocities(strong_bm_sample, f, g, dx, dy)
speed_bm = np.sqrt(u_bm**2 + v_bm**2)
vort_bm, strain_bm = calculate_flow_properties(u_bm, v_bm, dx, dy)


# Calculate for UBM
u_ubm, v_ubm = calculate_geostrophic_velocities(strong_ubm_sample, f, g, dx, dy)
speed_ubm = np.sqrt(u_ubm**2 + v_ubm**2)
vort_ubm, strain_ubm = calculate_flow_properties(u_ubm, v_ubm, dx, dy)

# Normalize vorticity and strain by f
vort_ssh_normalized = vort_ssh / abs(f)
strain_ssh_normalized = strain_ssh / abs(f)
vort_bm_normalized = vort_bm / abs(f)
strain_bm_normalized = strain_bm / abs(f)
vort_ubm_normalized = vort_ubm / abs(f)
strain_ubm_normalized = strain_ubm / abs(f)

# Create a figure with 3x6 subplots
fig, axes = plt.subplots(3, 6, figsize=(24, 12))

# Define custom colormaps for each variable type
scalar_cmap = 'RdBu_r'        # For SSH, BM and UBM
velocity_cmap = 'magma'       # For velocity components
speed_cmap = cmo.ice          # For speed
vorticity_cmap = cmo.thermal    # For vorticity
strain_cmap = 'magma'          # For strain magnitude

# Row 0: SSH fields
# SSH
im00 = axes[0, 0].imshow(strong_ssh_sample, cmap=scalar_cmap, vmin=-1, vmax=1)
axes[0, 0].set_title(r'$\eta_{SSH}$', fontsize=14)
fig.colorbar(im00, ax=axes[0, 0], shrink=0.8, label='(m)')

# u_ssh
u_ssh_max = max(abs(u_ssh.min()), abs(u_ssh.max()))
im01 = axes[0, 1].imshow(u_ssh, cmap=velocity_cmap, vmin=-1, vmax=1)
axes[0, 1].set_title(r'$u_{SSH}$', fontsize=14)
fig.colorbar(im01, ax=axes[0, 1], shrink=0.8, label='(m/s)')

# v_ssh
v_ssh_max = max(abs(v_ssh.min()), abs(v_ssh.max()))
im02 = axes[0, 2].imshow(v_ssh, cmap=velocity_cmap, vmin=-1, vmax=1)
axes[0, 2].set_title(r'$v_{SSH}$', fontsize=14)
fig.colorbar(im02, ax=axes[0, 2], shrink=0.8, label='(m/s)')

# speed_ssh
im03 = axes[0, 3].imshow(speed_ssh, cmap=speed_cmap, vmin=0, vmax=1)
axes[0, 3].set_title(r'$s_{SSH} = \sqrt{u_{SSH}^2 + v_{SSH}^2}$', fontsize=14)
fig.colorbar(im03, ax=axes[0, 3], shrink=0.8, label='(m/s)')

# vorticity_ssh (normalized)
vort_ssh_norm_max = max(abs(vort_ssh_normalized.min()), abs(vort_ssh_normalized.max()))
im04 = axes[0, 4].imshow(vort_ssh_normalized, cmap=vorticity_cmap, vmin=-vort_ssh_norm_max, vmax=vort_ssh_norm_max)
axes[0, 4].set_title(r'$\zeta_{SSH}/|f| = (\nabla \times \vec{u}_{SSH})/|f|$', fontsize=14)
fig.colorbar(im04, ax=axes[0, 4], shrink=0.8, label='(dimensionless)')

# strain_ssh (normalized)
strain_ssh_norm_max = max(abs(strain_ssh_normalized.min()), abs(strain_ssh_normalized.max()))
im05 = axes[0, 5].imshow(strain_ssh_normalized, cmap=strain_cmap, vmin=0, vmax=strain_ssh_norm_max)
axes[0, 5].set_title(r'$\sigma_{SSH}/|f| = \sqrt{\sigma_n^2 + \sigma_s^2}/|f|$', fontsize=14)
fig.colorbar(im05, ax=axes[0, 5], shrink=0.8, label='(dimensionless)')

# Row 1: BM fields
# BM
im10 = axes[1, 0].imshow(strong_bm_sample, cmap=scalar_cmap, vmin=-1, vmax=1)
axes[1, 0].set_title(r'$\eta_{BM}$', fontsize=14)
fig.colorbar(im10, ax=axes[1, 0], shrink=0.8, label='(m)')

# u_bm
u_bm_max = max(abs(u_bm.min()), abs(u_bm.max()))
im11 = axes[1, 1].imshow(u_bm, cmap=velocity_cmap, vmin=-1, vmax=1)
axes[1, 1].set_title(r'$u_{BM}$', fontsize=14)
fig.colorbar(im11, ax=axes[1, 1], shrink=0.8, label='(m/s)')

# v_bm
v_bm_max = max(abs(v_bm.min()), abs(v_bm.max()))
im12 = axes[1, 2].imshow(v_bm, cmap=velocity_cmap, vmin=-1, vmax=1)
axes[1, 2].set_title(r'$v_{BM}$', fontsize=14)
fig.colorbar(im12, ax=axes[1, 2], shrink=0.8, label='(m/s)')

# speed_bm
im13 = axes[1, 3].imshow(speed_bm, cmap=speed_cmap, vmin=0, vmax=1)
axes[1, 3].set_title(r'$s_{BM} = \sqrt{u_{BM}^2 + v_{BM}^2}$', fontsize=14)
fig.colorbar(im13, ax=axes[1, 3], shrink=0.8, label='(m/s)')

# vorticity_bm (normalized)
vort_bm_norm_max = max(abs(vort_bm_normalized.min()), abs(vort_bm_normalized.max()))
im14 = axes[1, 4].imshow(vort_bm_normalized, cmap=vorticity_cmap, vmin=-vort_bm_norm_max, vmax=vort_bm_norm_max)
axes[1, 4].set_title(r'$\zeta_{BM}/|f| = (\nabla \times \vec{u}_{BM})/|f|$', fontsize=14)
fig.colorbar(im14, ax=axes[1, 4], shrink=0.8, label='(dimensionless)')

# strain_bm (normalized)
strain_bm_norm_max = max(abs(strain_bm_normalized.min()), abs(strain_bm_normalized.max()))
im15 = axes[1, 5].imshow(strain_bm_normalized, cmap=strain_cmap, vmin=0, vmax=strain_bm_norm_max)
axes[1, 5].set_title(r'$\sigma_{BM}/|f| = \sqrt{\sigma_n^2 + \sigma_s^2}/|f|$', fontsize=14)
fig.colorbar(im15, ax=axes[1, 5], shrink=0.8, label='(dimensionless)')

# Row 2: UBM fields
# UBM
im20 = axes[2, 0].imshow(strong_ubm_sample, cmap=scalar_cmap, vmin=-0.01, vmax=0.01)
axes[2, 0].set_title(r'$\eta_{UBM}$', fontsize=14)
fig.colorbar(im20, ax=axes[2, 0], shrink=0.8, label='(m)')

# u_ubm
u_ubm_max = max(abs(u_ubm.min()), abs(u_ubm.max()))
im21 = axes[2, 1].imshow(u_ubm, cmap=velocity_cmap, vmin=-0.15, vmax=0.15)
axes[2, 1].set_title(r'$u_{UBM}$', fontsize=14)
fig.colorbar(im21, ax=axes[2, 1], shrink=0.8, label='(m/s)')

# v_ubm
v_ubm_max = max(abs(v_ubm.min()), abs(v_ubm.max()))
im22 = axes[2, 2].imshow(v_ubm, cmap=velocity_cmap, vmin=-0.15, vmax=0.15)
axes[2, 2].set_title(r'$v_{UBM}$', fontsize=14)
fig.colorbar(im22, ax=axes[2, 2], shrink=0.8, label='(m/s)')

# speed_ubm
im23 = axes[2, 3].imshow(speed_ubm, cmap=speed_cmap, vmin=0, vmax=0.2)
axes[2, 3].set_title(r'$s_{UBM} = \sqrt{u_{UBM}^2 + v_{UBM}^2}$', fontsize=14)
fig.colorbar(im23, ax=axes[2, 3], shrink=0.8, label='(m/s)')

# vorticity_ubm (normalized)
vort_ubm_norm_max = max(abs(vort_ubm_normalized.min()), abs(vort_ubm_normalized.max()))
im24 = axes[2, 4].imshow(vort_ubm_normalized, cmap=vorticity_cmap, vmin=-vort_ubm_norm_max, vmax=vort_ubm_norm_max)
axes[2, 4].set_title(r'$\zeta_{UBM}/|f| = (\nabla \times \vec{u}_{UBM})/|f|$', fontsize=14)
fig.colorbar(im24, ax=axes[2, 4], shrink=0.8, label='(dimensionless)')

# strain_ubm (normalized)
strain_ubm_norm_max = max(abs(strain_ubm_normalized.min()), abs(strain_ubm_normalized.max()))
im25 = axes[2, 5].imshow(strain_ubm_normalized, cmap=strain_cmap, vmin=0, vmax=strain_ubm_norm_max)
axes[2, 5].set_title(r'$\sigma_{UBM}/|f| = \sqrt{\sigma_n^2 + \sigma_s^2}/|f|$', fontsize=14)
fig.colorbar(im25, ax=axes[2, 5], shrink=0.8, label='(dimensionless)')

# Turn off axis ticks for all subplots
for ax in axes.flat:
    ax.set_xticks([])
    ax.set_yticks([])

# Fix layout to prevent overlap
plt.tight_layout()

# Add a title to the whole figure
plt.suptitle('SSH, Balanced Motion (BM) and Unbalanced Motion (UBM) Fields', 
             fontsize=16, y=1.02)

# Add text with Coriolis parameter value
plt.figtext(0.5, -0.02, f'Coriolis parameter $f = {f:.2e}$ s$^{{-1}}$ (Agulhas region, approx. 40°S)', 
            ha='center', fontsize=12)

# Save figure with high resolution
plt.savefig('figures/strong_ssh.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Define a function to calculate the PSD if it's not already defined
def calculate_psd_km(field, dx_km=1.5):
    """
    Calculate the 2D Power Spectral Density and convert to 1D radial average.
    
    Parameters:
    -----------
    field : 2D numpy array
        The input field (e.g., SSH, BM, UBM)
    dx_km : float
        Grid spacing in kilometers
        
    Returns:
    --------
    psd_1d : numpy array
        1D radially averaged power spectrum
    freq_r : numpy array
        Corresponding radial frequencies in cycles/km
    """
    # Get dimensions
    ny, nx = field.shape
    
    # Apply Hann window to reduce edge effects
    window = np.outer(np.hanning(ny), np.hanning(nx))
    field_windowed = field * window
    
    # Compute 2D FFT and shift zero frequency to center
    fft2 = np.fft.fft2(field_windowed)
    fft2_shifted = np.fft.fftshift(fft2)
    
    # Calculate the Power Spectral Density
    psd2d = np.abs(fft2_shifted)**2
    
    # Normalize by number of points and grid spacing
    psd2d = psd2d / (nx * ny) * (dx_km**2)
    
    # Create frequency grids
    freq_x = np.fft.fftshift(np.fft.fftfreq(nx, d=dx_km))
    freq_y = np.fft.fftshift(np.fft.fftfreq(ny, d=dx_km))
    
    # Create meshgrid for radial binning
    kx, ky = np.meshgrid(freq_x, freq_y)
    k_rad = np.sqrt(kx**2 + ky**2)
    
    # Create radial bins
    max_freq = np.max(k_rad)
    n_bins = min(nx, ny) // 2
    freq_bins = np.linspace(0, max_freq, n_bins+1)
    
    # Calculate bin centers
    freq_r = (freq_bins[1:] + freq_bins[:-1]) / 2
    
    # Initialize output array
    psd_1d = np.zeros(n_bins)
    
    # Radial binning
    for i in range(n_bins):
        bin_mask = (k_rad >= freq_bins[i]) & (k_rad < freq_bins[i+1])
        if np.any(bin_mask):
            psd_1d[i] = np.mean(psd2d[bin_mask])
    
    return psd_1d, freq_r

# Create the figure
plt.figure(figsize=(10, 8))

# Extract fields for this sample
ssh = strong_ssh_sample  # Use the variable you already defined
bm = strong_bm_sample    # Use the variable you already defined
ubm = strong_ubm_sample  # Use the variable you already defined

# Calculate PSDs
psd_ssh, freq_r_ssh = calculate_psd_km(ssh)
psd_bm, freq_r_bm = calculate_psd_km(bm)
psd_ubm, freq_r_ubm = calculate_psd_km(ubm)

# Plot on log-log scale
plt.loglog(freq_r_ssh, psd_ssh, label='SSH', linewidth=2, color='blue')
plt.loglog(freq_r_bm, psd_bm, label='BM', linewidth=2, color='orange')
plt.loglog(freq_r_ubm, psd_ubm, label='UBM', linewidth=2, color='green')

# Add reference wavelengths
wavelengths = [100, 70, 50, 30, 20, 10]  # in km
for wl in wavelengths:
    freq = 1.0 / wl
    plt.axvline(x=freq, linestyle=':', color='gray', alpha=0.5)
    plt.text(freq, plt.ylim()[0]*1.2, f'{wl} km', 
            rotation=90, va='bottom', ha='center', fontsize=10, alpha=0.7)

# Add reference slopes for spectral cascades
# For k^-5 (Surface QG turbulence)
k_ref = np.array([0.01, 0.1])
E_ref_k5 = 1e-2 * (k_ref/k_ref[0])**(-5)
plt.loglog(k_ref, E_ref_k5, 'k--', linewidth=1.5, label=r'$k^{-5}$')

# For k^-4 (Frontogenesis)
E_ref_k4 = 1e-3 * (k_ref/k_ref[0])**(-4)
plt.loglog(k_ref, E_ref_k4, 'k:', linewidth=1.5, label=r'$k^{-4}$')

# Set axis labels and title
plt.xlabel(r'Wavenumber (cycles km$^{-1}$)', fontsize=14)
plt.ylabel(r'PSD (m$^{2}$ / cycles km$^{-1}$)', fontsize=14)
plt.title('Strong SSH case', fontsize=16)
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend(frameon=True, fontsize=12, loc='best')

# Add annotation about the Coriolis parameter
plt.figtext(0.5, 0.01, f'Coriolis parameter $f = {f:.2e}$ s$^{{-1}}$ (Agulhas region)', 
            ha='center', fontsize=12)

# Adjust spacing and save
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)
plt.savefig('figures/psd_strong_ssh.png', dpi=300, bbox_inches='tight')
plt.show()
```
